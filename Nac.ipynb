```python
# how to create code to read image and identify the object in the image using python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# read the image
img = cv2.imread('image.jpg')
# convert the image to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# show the image
plt.imshow(img)
plt.show()
# convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# show the grayscale image
plt.imshow(gray, cmap='gray')
plt.show()
# create a binary image
ret, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)
# show the binary image
plt.imshow(binary, cmap='gray')
plt.show()
# find the contours
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# draw the contours
img_copy = img.copy()
img_contours = cv2.drawContours(img_copy, contours, -1, (0, 255, 0), 3)
# show the image with contours
plt.imshow(img_contours)
plt.show()

# create a function to find the contours
def get_contours(img):
    # convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # create a binary image
    ret, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)
    # find the contours
    contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    return contours

# create a function to draw the contours
def draw_contours(img, contours):
    # draw the contours
    img_copy = img.copy()
    img_contours = cv2.drawContours(img_copy, contours, -1, (0, 255, 0), 3)
    return img_contours

# find the contours
contours = get_contours(img)
# draw the contours
img_contours = draw_contours(img, contours)
# show the image with contours
plt.imshow(img_contours)
plt.show()



  
```
